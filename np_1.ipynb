```python
ctd=['id','sg','al','sc','htn','cad']
```

```python
kai.dropna(inplace=True)
```

```python
label_encoder = LabelEncoder()
kai['rbc']= label_encoder.fit_transform(kai['rbc'])
kai['pc']= label_encoder.fit_transform(kai['pc'])
kai['pcc']= label_encoder.fit_transform(kai['pcc'])
kai['ba']= label_encoder.fit_transform(kai['ba'])
kai['dm']= label_encoder.fit_transform(kai['dm'])
kai['classification']=label_encoder.fit_transform(kai['classification'])
kai['pe']=label_encoder.fit_transform(kai['pe'])
kai['ane']=label_encoder.fit_transform(kai['ane'])
kai['appet']=label_encoder.fit_transform(kai['appet'])
```
